# Data loading

````python
def prepare_sequence(seq, to_ix):
    """
    Convert a sequence of words to a sequence of indices based on a provided mapping.

    Parameters:
    seq (list): A list of words.
    to_ix (dict): A dictionary mapping words to unique indices.

    Returns:
    torch.tensor: A tensor of indices corresponding to the words in the sequence.
    """
    idxs = [to_ix[w] for w in seq]
    return torch.tensor(idxs, dtype=torch.long)
````

````python
def read_dataset(filename, splitter="\t"):
    """
    Read a dataset from a file and return a list of (sentence, tags) tuples.

    Parameters:
    filename (str): Path to the dataset file.
    splitter (str): The character used to split words from their tags in the dataset.

    Returns:
    list: A list of tuples, where each tuple contains a list of words (sentence) and their corresponding tags.
    """
    # Function implementation...

def read_dataset(filename, splitter="\t"):
    data = []
    sentence = []
    tags = []
    with open(filename) as f:
        for line in f:
            if not line.isspace():
                word, tag = line.split(splitter)
                sentence.append(word)
                tags.append(tag.strip())
            else:
                data.append((sentence, tags))
                sentence = []
                tags = []
    return data
````

````python

def load_pretrained_embeddings(word_to_ix, embeddings_file, embedding_dim):

    # Initialize an embeddings matrix with zeros
    embeddings_matrix = np.zeros((len(word_to_ix), embedding_dim))

    # Read the pre-trained embeddings file
    with open(embeddings_file, 'r', encoding='utf-8') as file:
        for line in file:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            # Add the word's vector to the matrix if the word is in our vocabulary
            if word in word_to_ix:
                embeddings_matrix[word_to_ix[word]] = vector

    return embeddings_matrix
````

# Tagger

###  Create Tagger Class

````python

class SRLTagger(nn.Module): # Using LSTMT as underlying RNN
    def __init__(self, embedding_dim, hidden_dim, vocab_size, tagset_size, ):
        super(SRLTagger, self).__init__()
        self.hidden_dim = hidden_dim
        self.word_embeddings = nn.Embedding(vocab_size, embedding_dim)
        self.rnn = nn.LSTM(embedding_dim, hidden_dim, batch_first=True)
        self.hidden2tag = nn.Linear(hidden_dim, tagset_size)
        self.crf = CRF(tagset_size, batch_first=True)

    def forward(self, sentence):
        embeds = self.word_embeddings(sentence)
        rnn_out, _ = self.rnn(embeds)
        tag_space = self.hidden2tag(rnn_out)
        return tag_space

    def calculate_loss(self, sentences, tags):
        features = self.forward(sentences)
        return -self.crf(features, tags)  # Negative log-likelihood

    def predict(self, sentences):
        features = self.forward(sentences)
        return self.crf.decode(features)

    def init_embeddings(self, embeddings_matrix):
        self.word_embeddings.weight = nn.Parameter(torch.tensor(embeddings_matrix, dtype=torch.float))
````

# Training

````python


model_params = {
    'embedding_dim': embeddings_matrix.shape[1],
    'hidden_dim': HIDDEN_DIM,
    'vocab_size': len(word_to_ix),
    'tagset_size': len(tag_to_ix),
    'rnn_type': 'LSTM'
}


model = SRLTagger(**model_params)
model.init_embeddings(embeddings_matrix)
````

````python



no_epochs = 30
no_folds = 4
kfold = KFold(n_splits=no_folds, shuffle=True, random_state=42)


learning_rates = [0.1 , 0.01 , 0.001]
hidden_dims = [128, 256 , 512]
tolerance = 5  # Min num of epochs to run before potentially stopping after
#inporovement check


best_valid_loss = float('inf')
best_lr = None


for hidden_dim in hidden_dims:
    model_params['hidden_dim'] = hidden_dim
    for lr in learning_rates:
        fold_valid_losses = []

        for fold, (train_ids, valid_ids) in enumerate(kfold.split(training_data)):
            print(f"Training on fold {fold+1} with learning rate {lr}, hidden dimension: {hidden_dim}")

            # Split data into training and validation
            train_subsampler = torch.utils.data.SubsetRandomSampler(train_ids)
            valid_subsampler = torch.utils.data.SubsetRandomSampler(valid_ids)

            # Initialize model
            model = SRLTagger(**model_params)
            model.init_embeddings(embeddings_matrix)
            optimizer = optim.SGD(model.parameters(), lr=lr)

            current_valid_loss = float('inf')
            patience_counter = 0  # Counter for early stopping

            # Training loop
            for epoch in tqdm(range(num_epochs), desc=f"Fold {fold+1}, Epoch"):
                model.train()  # Training mode
                total_loss = 0


                for sentence, tags in [training_data[i] for i in train_ids]:
                    model.zero_grad()
                    sentence_in = prepare_sequence(sentence, word_to_ix).unsqueeze(0)
                    targets = prepare_sequence(tags, tag_to_ix).unsqueeze(0)
                    loss = model.calculate_loss(sentence_in, targets)
                    loss.backward()
                    optimizer.step()
                    total_loss += loss.item()

                # Validation phase
                model.eval()
                valid_loss = 0
                with torch.no_grad():
                    for sentence, tags in [training_data[i] for i in valid_ids]:
                        sentence_in = prepare_sequence(sentence, word_to_ix).unsqueeze(0)
                        targets = prepare_sequence(tags, tag_to_ix).unsqueeze(0)
                        valid_loss += model.calculate_loss(sentence_in, targets).item()

                avg_valid_loss = valid_loss / len(valid_ids)

                # Early stopping check
                if avg_valid_loss < current_valid_loss:
                    current_valid_loss = avg_valid_loss
                    patience_counter = 0
                else:
                    patience_counter += 1

                if patience_counter >= patience:
                    print(f"Stopping early at epoch {epoch + 1} for learning rate {lr}")
                    break

            fold_valid_losses.append(current_valid_loss)


        avg_lr_valid_loss = sum(fold_valid_losses) / len(fold_valid_losses)
        print(f"Average validation loss for learning rate {lr}: {avg_lr_valid_loss:.4f}")

        # Update the best learning rate based on average validation loss
        if avg_lr_valid_loss < best_valid_loss:
            best_valid_loss = avg_lr_valid_loss
            best_lr = lr
            best_hidden_dim = hidden_dim

print(f"Best Hyperparameters - Learning Rate: {best_lr}, Hidden Dimension: {best_hidden_dim}, Validation Loss: {best_valid_loss:.4f}")
````
